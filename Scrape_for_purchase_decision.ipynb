## Filter best offers

```python
prod_var_list = (set(final_data['prod_var']))
```

```python
sku_list = []
product_list = []
variant_list = []
rank_list = []
location_list = []
pp_list = []
pq_list = []
tp_cost_list = []
fq_rm_list = []

for pv in prod_var_list:
  temp_data = final_data[final_data['prod_var']==pv]
  max_rank = max(temp_data['Rank_Order'])
  best_skus = temp_data[temp_data['Rank_Order']==max_rank]
  lowest_price = best_skus[best_skus['Puchase Price']==min(best_skus['Puchase Price'])]

  for key in lowest_price['key']:
    sku_list.append(key)
  for prod in lowest_price['Product']:
    product_list.append(prod)
  for var in lowest_price['Variant']:
    variant_list.append(var)
  for rank in lowest_price['Rank_Order']:
    rank_list.append(rank)
  for loc in lowest_price['Location']:
    location_list.append(loc)
  for pp in lowest_price['Puchase Price']:
    pp_list.append(pp)
  for pq in lowest_price['Purchase Quantity']:
    pq_list.append(pq)
  for rs in lowest_price['final_qunatity_removing_stock']:
    fq_rm_list.append(rs)
  for tc in lowest_price['Total Purchase Cost']:
    tp_cost_list.append(tc)
```

```python
decision_data = pd.DataFrame({
  'key': sku_list,
  'product': product_list,
  'variant': variant_list,
  'rank': rank_list,
  'location': location_list,
  'price': pp_list,
  'quantity': pq_list,
  'quantity_removing_stock': fq_rm_list,
  'total_purchase_cost': tp_cost_list
})
```
